```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
list_url = 'https://www.dictionary.com/list/{}'
alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
```

```python
session = requests.session()
```

```python
url = list_url.format(alphabet[0])
print(url)
html_list = session.get(url)
list_soup = BeautifulSoup(html_list.content,'html.parser')
```

```python
def get_words_from_html_list(html_list):
    words = []
    for word_item in html_list:
        anchor_el = word_item.find('a')
        words.append((anchor_el.text,anchor_el['href']))
    return words

def get_words(letter):
    words = []
    url = list_url.format(letter)
    print(url)
    html_list = session.get(url)
    html_soup = BeautifulSoup(html_list.content, 'html.parser')
    try:
        num_pages = int(html_soup.find('ol',{'class':'css-8058rk-PaginatorList'}).findAll('a')[-1]['data-page'])
    except:
        num_pages = 0
    print('Num pages:',num_pages)
    words += get_words_from_html_list(html_soup.find('ul',{'class':'css-1y59cbu'}))
    for i in range(2, num_pages+1):
        new_page = url + '/' + str(i)
        if i == 2:
            print(new_page)
        html_list = session.get(new_page)
        html_soup = BeautifulSoup(html_list.content, 'html.parser')
        words += get_words_from_html_list(html_soup.find('ul',{'class':'css-1y59cbu'}))
    return words
```

## Getting list of words in link

```python
# save each letter in text file
for letter in alphabet:
    start_time = time.time()
    print(' {} '.format(letter).center(20,'-'))
    letter_words = get_words(letter)
    df_words = pd.DataFrame(letter_words, columns=['word','link'])
    path = '../dictionary/letter_lists/{}.csv'.format(letter)
    df_words.to_csv(path,index=False)
    end_time = time.time()
    total = end_time - start_time
    print('{} seconds'.format(round(total,3)))
```
